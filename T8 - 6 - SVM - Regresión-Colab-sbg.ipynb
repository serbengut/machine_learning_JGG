# SVM para Regresión

```python
# no vamos a clasificar ahora, si no que se trata de un problema de regresion
# crearemos un dataset aleatorio y vamos a intentar buscar un soporte vectorial que nos de un modelo de regresión (que se ajuste lo mejor posible a los puntos)
# normalmente, para tener una buena visualización y para entender como funciona, lo suyo es usar un solo rasgo como entrada de datos con nuestro ejemplo
# vamos a intentar que los puntos no sigan una distribución lineal, porque si no ya lo tendríamos
# para ello, vamos a usar la función seno que no se puede ajustar con ningun tipo de recta
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.sort(5*np.random.rand(200,1),axis=0)
Y = np.sin(X).ravel()
plt.scatter(X,Y, color="darkorange", label="data")
```

```python
X = np.sort(5*np.random.rand(200,1),axis=0)
Y = np.sin(X).ravel()
Y[::5] += 3*(0.5 - np.random.rand(40)) #creo los puntos aleatorios de 40 numeros que se saldran del seno
```

```python
plt.scatter(X,Y, color="darkorange", label="data")
```

```python
from sklearn.svm import SVR
```

```python
C=1e3
svr_lin = SVR(kernel="linear", C=C)
svr_rbf = SVR(kernel="rbf", C=C, gamma=0.1)
svr_pol = SVR(kernel="poly", C=C, degree=3)
```

```python
y_lin = svr_lin.fit(X,Y).predict(X)
y_rbf = svr_rbf.fit(X,Y).predict(X)
y_pol = svr_pol.fit(X,Y).predict(X)
```

```python
lw = 2
plt.figure(figsize=(16,9))
plt.scatter(X,Y,color="darkorange", label ="data")
plt.plot(X,y_lin, color="navy", lw = lw, label = "SVM Lineal")
plt.plot(X,y_rbf, color="g", lw=lw, label="SVM Radial")
plt.plot(X,y_pol, color="cornflowerblue", label="SVM Polinómico")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Support Vector Regression")
plt.legend()
plt.show()
```
