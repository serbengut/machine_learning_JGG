```python
# Diferentes reglas para calcular las divisiones (bins) del histograma

import numpy as np
def sturges(data):
   # para numero datos mayores de 30
   num_data = len(data)
   num_bins = 1 + int(np.log2(num_data))
   print(num_bins)
   return num_bins

def freedman_diaconis(data):
   num_data = len(data)
   irq = np.percentile(data, 75) - np.percentile(data, 25)
   bin_width = 2 * irq / np.power(num_data, 1/3)
   num_bins = int((np.max(data) -  np.min(data)) / bin_width)  + 1
   print(num_bins)
   return num_bins

def scott(data):
   num_data = 1/(len(data))**(1/3)    
   num_bins = int(np.ceil(3.5*(np.std(data))*num_data)) #np.std standard deviation
   print(num_bins)
   return num_bins

def raiz(data):
   num_data = len(data)
   num_bins = int(np.ceil(np.sqrt(num_data)))    
   print(num_bins)
   return num_bins
```

```python
sturges(data)
```

```python
freedman_diaconis(data)
```

```python
scott(data)
```

```python
raiz(data)
```

### Boxplot, diagrama de caja y bigotes

```python
plt.boxplot(data["Day Calls"])
plt.ylabel("N√∫mero de llamadas diarias")
plt.title("Boxplot de las llamadas diarias")
```

```python
# la caja empieza por abajo en el percentil 25 (es decir, por debajo est√° el 25% de los datos m√°s peque√±os).
# la caja acaba por arriba en el percentil 75 (es decir, por encima de ella est√° el 25% de los datos m√°s grandes)
# la raya naranja es la mediana o el percentil 50
# la moda es el valor que m√°s sale
# la media es la media de todos los valores
# rango interquatilico (irq): diferencia entre percentil 75 y percentil 25 (es el tama√±o de la caja)
# los bigotes se colocan a 1.5 veces el irq
# m√°s all√° de los bigotes los datos se representan con bolitas
# las bolitas son los outlier
```
