```python
import tensorflow as tf
```

```python
# normalmente con tf trabajamos con array
```

```python
x1 = tf.constant([1,2,3,4,5])   #tf.constant crea una constante
x2 = tf.constant([6,7,8,9,10])
```

```python
res = tf.multiply(x1,x2)
print(res)
```

```python
# parece que esto no funciona, ya nos da el resultado de multiplicar x1 y x2 arriba [6,14,24,36,50], por lo que no necesitamos esto....

sess = tf.Session()
print(sess.run(res))
sess.close()
```

```python
# parece que esto no funciona, ya nos da el resultado de multiplicar x1 y x2 arriba [6,14,24,36,50], por lo que no necesitamos esto....


with tf.Session() as sess:
    output = sess.run(res)
    print(output)
```
