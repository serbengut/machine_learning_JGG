```python
import numpy as np
import pandas as pd
import math
import random
```

```python
# vamos a generar dos listas con estas opciones:
gender = ["Male" , "Female"]
tour = ["PGA", "DPWorld", "Asian"]
```

```python
n=150

gender_data=[]
tour_data=[]

for i in range(0,n):
    gender_data.append(np.random.choice(gender))
    tour_data.append(np.random.choice(tour))
```

```python
# no se puede hacer head porque es una lista, no un dataframe
gender_data.head()
```

```python
gender_data[0:10]
```

```python
tour_data[0:10]
```

```python
# vamos a generar nuevas variables con una distribución normal:

#Z -> N(0,1)
#N(m, s) -> m + s * Z

shots_per_round=72 + 3*np.random.randn(n)
```

```python
shots_per_round[0:10]
```

```python
birdies_per_round_raw = 2 + 1*np.random.randn(n)
birdies_per_round_rounded = ["%.0f" % elem for elem in birdies_per_round_raw]
birdies_per_round=[int(x) for x in birdies_per_round_rounded]
for x in range(len(birdies_per_round)):
    if birdies_per_round[x]<0:
        birdies_per_round[x]=0
```

```python
birdies_per_round[0:10]
```

```python
income_per_tournament_raw=25000 + 5000*np.random.randn(n)
income_per_tournament_rounded = ["%.2f" % elem for elem in income_per_tournament_raw]
income_per_tournament = [round(float(s),2) for s in income_per_tournament_rounded]
```

```python
income_per_tournament[0:10]
```

```python
data_golf = pd.DataFrame (
    {
      "Gender": gender_data,
     "Tour": tour_data,
     "Shots Per Round": shots_per_round,
     "Birdies Per Round": birdies_per_round,
     "Income": income_per_tournament  
    }
)
```
