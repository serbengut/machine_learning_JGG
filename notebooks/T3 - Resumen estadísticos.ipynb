## 2. Medidas de centralización

Nos sirve para ver cómo se sitúan los datos. Son la **media**, **mediana**, **peprcentiles** y **moda**.

Vamos a generar una lista de números

```python
# generamos números aleatorios
import random
random.seed(0)
A = (random.sample(range(100000),60))
A
```

```python
# calculo la media
mean = np.mean(A)
mean
```

```python
# calculo la mediana
np.median(A)
```

```python
# calculo la moda
from scipy import stats
stats.mode(A)
```

```python
# calculo percentil: 
# percentil 100 = p1
# 0 = po
#0,25 = 1er cuartil
#0.75 = 3er cuartil
np.percentile(A,25)
```

## 3. Medidas de dispersión

La varianza y desviación típica, nos indica si los valores se desplazan mucho o poco con respecto de la media.

La **varianza** es cómo se aleja cada valor de la media. La varianza eleva los valores al cuadrado... nos introduce en una nueva dimensión... puede no tener sentido.

La **desviación típica** es la raiz cuadrada de la varianza. Con la desviación típica volvemos a la dimensión original.

**Coeficiente de variación**: nos mide la variabilidad relativa entre la desviación típica y entre la media

```python
# calculamos la varianza
np.var(A)
```

```python
# calculamos la desviación típica

std=np.std(A)
std
```

```python
# coeficiente de varianza: std/mean*100
# variabilidad relativa entre la media y la std, si hay mucha variabilidad será grande el coeficiente
std/mean*100
```

## 4. Medidas de asimetría

Momento de orden r, respecto a la media. El momento de orden r son los momentos de distribución respecto a la media

```python
# para modelos de asimetría
import scipy.stats as sp
# para incluir imagenes en local
from IPython.display import Image
```

```python
Image(filename="/content/drive/MyDrive/Python Machine Learning JGG/notebooks/resources/asimetría de fisher.png")
```

**Asimetría de Fisher**

- Si el coeficiente es = 0 ; significa que nuestra función es perfectamente simétrica, se distribuye igual, por ejemplo, la distribución normal. Es raro que salga 0.

- Si el coeficiente es > 0 ; significa que cuánto más positivo es este valor, más desplazada está la distribución hacia la izquierda, de modo que tenemos una asimetría positiva, nos queda la media muy por encima de la distribución

- Si el coeficiente es < 0 ; significa que cuánto más negativo es este valor, más desplazada está la distribución hacia la derecha, de modo que tenemos una asimetría negativa, nos queda la media muy por debajo de la distribución

```python
# 3er Moment: Asimetría de Fisher
Asimetria = sp.skew(A)
Asimetria
```

**Curtosis**

- = 0 Mesocúrtica: Distribución perfecta, asemejada a la distribución normal en forma, no en valores. Está compensado tanto el centro como las colas. 

- '> 0 Leptocúrtica: Distribución donde se le concentran mucho los datos en el valor central, y apenas tiene cola.

- < 0 Platicúrtica: Distribución donde hay pocos valores que se concentran respecto al valor central (media) y hay muchos que aparecen hacia las colas. Se concentran más en los laterales. Existe valor central, pero también hay mucha presencia de colas directamente en la distribución de nuestros datos

```python
# Curtosis
Image(filename="/content/drive/MyDrive/Python Machine Learning JGG/notebooks/resources/curva_curtosis.jpg")
```

```python
# 4th Moment: Curtosis
curtosis=sp.kurtosis(A)
curtosis
```
