```python
from sklearn.feature_selection import RFE
from sklearn.svm import SVR
import pandas as pd
import numpy as np
```

# Regresión lineal en Python
## El paquete scikit-learn para regresión lineal y la selección de rasgos

## OJO

En la versión 3.9 de python, al menos a la fecha la función sklearn.feature_selection  RFE ha sufrido un cambio en el segundo parámetro que genera un fallo, pasa de
RFE(estimator,  2, step=1)
a
RFE(estimator,  n_features_to_select=2, step=1)
para que funcione correctamente.

```python
# RFE: recursive feature elimination. es similar a eliminar cada vez una variable. simplemente le tenemos que decir cuántas variables queremos tener. él solo va poniendo y 
# quitando hasta quedarse con el número de vbles que le hemos indicado
```

```python
data = pd.read_csv("/content/drive/MyDrive/Python Machine Learning JGG/datasets/ads/Advertising.csv")
```

```python
data.head()
```

```python
# nombro las columnas de predicción:

feature_cols = ["TV", "Radio", "Newspaper"]
```

```python
# separo en dos datasets:
# X: el que se lleva las vbles candidatas a vbles. predictoras
# Y: la vble predecida

X = data[feature_cols]
Y = data["Sales"]
```

```python
estimator = SVR(kernel="linear") # dentro de RFE usamos lineal
selector = RFE(estimator, n_features_to_select=2, step=1) # el 2 significa que quiero quedarme con dos vbles predictoras; el step=1 es para que lo haga en un paso
selector = selector.fit(X,Y)
```

```python
# para ver la lista de las vbles con las que nos hemos quedado:

selector.support_
```

```python
# es decir, nos hemos quedadto con TV, Radio (true) y descartamos Newspaper (false)
# esto concluye, que la selección de variables que hicmos en el colab antes tiene sentido, ya que coincide con esto.

# también podemos pedirle un ranking de vbles:
```

```python
selector.ranking_
```

```python
# 1: significa que nos quedamos con ella (es más significativa que las de rango 2)
# 2: a medida que aumenta el rango, van perdiendo más significidad
```

```python
# ahora podemos hallar el modelo, o bien como se hizo en el colab anterior, o con scikit-learn:
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X_pred = X[["TV", "Radio"]]
```

```python
lm = LinearRegression()
lm.fit(X_pred,Y)
```

```python
lm.intercept_ # nos da la alfa
```

```python
lm.coef_ # nos da los coeficientes que van con TV y Radio
```

```python
lm.score(X_pred,Y) # nos da R2
```
