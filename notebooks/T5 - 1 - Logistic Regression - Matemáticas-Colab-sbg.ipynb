### Ratio de probabilidades
Cociente entre los casos de éxito sobre los de fracaso en el suceso estudiado y para cada grupo

```python
# en nuestro caso tenemos dos grupos: hombres y mujeres

display(Math(r'P_m = \ probabilidad\ de\ hacer\ compra\ sabiendo\ que\ es\ un\ hombre'))
print("")
display(Math(r'P_f = \ probabilidad\ de\ hacer\ compra\ sabiendo\ que\ es\ una\ mujer'))
print("")
display(Math(r"odds\in[0,+\infty]"))
print("")
display(Math(r'ods_{purchase,male} = \frac{P_m}{1-P_m} = \frac{N_{purchase,male}} {N_{No purchase,male}}  = \frac{N_{p,m}} {N_{\bar p,m}}'))
print("")
display(Math(r'ods_{purchase,female} = \frac{P_f}{1-P_f} = \frac{N_{purchase,female}} {N_{No purchase,female}} = \frac{N_{p,f}} {N_{\bar p,f}}'))
```

```python
# con la tabla de contingencia....

pm = 121/246
pf = 159/265

odds_m = pm / (1-pm) # es igual a: casos favorables / casos desfavorables: 121 / 125 = 0.96
odds_f = pf/(1-pf) # es igual a: casos favorables / casos desfavorables:  159 / 106 = 1.4999
```

```python
odds_m
```

```python
odds_f
```

* Si el ratio es superior a 1, es más probable el éxito que el fracaso. Cuanto mayor es el ratio, más probabilidad de éxito en nuestro suceso.

* Si el ratio es exactamente igual a 1, éxito y fracaso son equiprobables (p=0.5)

* Si el ratio es menor que 1, el fracaso es más probable que el éxito. Cuanto menor es el ratio, menor es la probabilidad de éxito del suceso.

```python
# en nuestro caso, las mujeres son más propensas a comprar que los hombres dado que odds_f = 1.5 y odds_m = 0.96
```

```python
display(Math(r"odds_{ratio} = \frac {odds_{purchase,male}}  {odds_{purchase,female}}"))
```

```python
odds_r = odds_m / odds_f
```

```python
odds_r
```

```python
1/odds_r# odds_f/odds_m
```

### La regresión logística desde la regresión lineal

```python
# Regresión LINEAL:

display(Math(r" y = \alpha + \beta \cdot x"))
display(Math(r"y \in[-\infty,+\infty]")) # x es CONTINUA
display(Math(r"x \in[-\infty,+\infty]")) # Y es CONTINUA
```

```python
# Regresión LOGÍSTICA:
```

```python
display(Math(r"Y \in\{0,1\}\ pero\ no\ sabenos\ cuánto\ vale")) # Y va a ser DISCRETA
display(Math(r"P \in[0,1]")) # la probabilidad p estará entre 0 y 1
display(Math(r"X \in[-\infty,+\infty]"))
```

```python
# en el caso de antes (compras según el género) solo teníamos una vble predictora: el género
# pero lo normal es que tengamos más vbles predictoras
# Encontrar el valor de y (0 ó 1) es complicado, no sabemos cuál es
# Buscamos la probabilidad P
```

```python
# podemos buscar la P con esta fórmula:
display(Math(r" P = \alpha + \beta \cdot X"))
```

P es la probabilidad condicionada de exito o de fracaso condicionada a la presencia de la variable X

```python
# habíamos dicho que en la regresión lineal, la x y la y iban de -infinito a +infinito
# pero en la logística, la P solo puede valer entre 0 y 1
# la X sí que puede ir de -inf a +inf (no hay problema)
```

```python
# para solucionar lo de la P hacemos esto:
display(Math(r'\frac{P}{1-P} = \alpha + \beta\cdot X')) # habíamos dicho que una probabilidad podía ir de 0 a infinito
print("")
display(Math(r"todo\ esto\ \in [0,+\infty]"))
```

```python
# ya hemos solucionado la parte derecha, nos falta la parte izquierda, ya que todavía parte de 0.
# si aplico un Ln al cociente de la izquierda, cubro todo el rango. es decir, si el rango de un número va de 0 a +infinito, entonces su logaritmo, aumenta su rango a [-infinito, + infinito]
```

```python
display(Math(r'ln(\frac{P}{1-P}) = \alpha + \beta\cdot X'))
print("")
display(Math(r"todo\ esto\ \in [-\infty,+\infty]"))
```

```python
# ya hemos conseguido que todo vaya de -inf a +inf
```

```python
display(Math(r'\begin{cases}\frac{P}{1-P}\in[0,1]\Rightarrow ln(\frac{P}{1-P})\in[-\infty,0]\\ \frac{P}{1-P}\in[1,+\infty]\Rightarrow ln(\frac{P}{1-P})\in[0, \infty]\end{cases}'))
```

```python
# con esto ya puedo sacar la parte derecha como si fuera una regresión lineal, pero necesito también sacar la probabilidad P
display(Math(r'ln(\frac{P}{1-P}) = \alpha + \beta\cdot X'))
```

```python
# despejo P de arriba:
    # para quitar el ln pongo "e" a cada lado, con lo que elimino en ln y a la derecha me queda e elevado a alpha + beta*X
```

```python
display(Math(r'ln(\frac{P}{1-P}) = \alpha + \beta\cdot X\Rightarrow'))
print("")
display(Math(r'\frac{P}{1-P} = e^{\alpha + \beta\cdot X}\Rightarrow'))
print("")
display(Math(r' P = \frac{e^{\alpha+\beta\cdot X}}{1+e^{\alpha+\beta\cdot X}}\Rightarrow'))
print("")
display(Math(r' P = \frac{1}{1+e^{-(\alpha+\beta\cdot X)}}'))
```

```python
# si al hacer el calculo de alfa+beta*x nos da un número muy muy muy grande, entonces e elevado a un número negativo muy muy grande es muy pequeño, entonces, la probabilidad tiende a 1, 
# por contra, si alfa+beta*x es muy muy negativo, entonces, P tiende a cero
```

* Si a+bX es muy pequeño (negativo), entonces P tiende a 0
* Si a+bX = 0, P = 0.5
* Si a+bX es muy grande (positivo), entonces P tiende a 1

### Para una Regresión logística múltiple

```python
display(Math(r' P = \frac{1}{1+e^{-(\alpha+\sum_{i=1}^n\beta_i\cdot x_i)}}'))
```

```python
display(Math(r' \vec{\beta} = (\beta_1,\beta_2,\cdots,\beta_n)'))
print("")
display(Math(r' \vec{X} = (x_1,x_2,\cdots,x_n)'))
print("")
display(Math(r' P = \frac{1}{1+e^{-(\alpha+\vec{\beta_i}\cdot \vec{X})}}'))
```
