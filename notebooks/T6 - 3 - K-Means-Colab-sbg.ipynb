```python
import numpy as np
```

```python
data = np.random.random(90).reshape(30,3) # saco 90 números aleatorios entre 0 y 1 y los reorganizo en 30 listas de 3 elementos cada una
data
```

```python
# recordamos que k-means tiene el problema de saber cuántos k hemos de tener
# entonces imaginamos que el analista, o el experto en markéting, decide que k=2
# seleccionamos dos obsesrvaciones al azar y esos serán los dos baricentros
```

```python
c1 = np.random.choice(range(len(data))) # elijo aleatoriamente un punto
c2 = np.random.choice(range(len(data))) # elijo aleatoriamente otro punto
clust_centers = ([data[c1], data[c2]])
clust_centers
```

```python
# el np.vstack lo pongo para poner un punto encima del otro, pero es lo mismo
clust_centers = np.vstack([data[c1], data[c2]])
clust_centers
```

```python
# ya tenemos los dos baricentros c1 y c2:
```

```python
from scipy.cluster.vq import vq
```

```python
clusters = vq(data, clust_centers)
clusters
```

```python
# el primer array nos da la información de qué puntos pertenecen a cada cluster
# el segundo array nos da la distancia de cada una de las 30 observaciones del baricentro del clustering final. es decir, cada numero representa la distancia de ese punto al baricentro del cluster que indique el primer array
```

```python
labels = clusters[0]
labels
```

```python
!pip install chart_studio
import chart_studio.plotly as py
import plotly.graph_objects as go 
import plotly.graph_objects as go
import plotly.offline as ply
```

```python
# esto no sé porqué lo ha puesto...
```

```python
x = []
y = []
z = []
x2 = []
y2 = []
z2 = []

for i in range(0, len(labels)):
    if(labels[i] == 0):
        x.append(data[i,0])
        y.append(data[i,1])
        z.append(data[i,2])
        
    else:
        x2.append(data[i,0])
        y2.append(data[i,1])
        z2.append(data[i,2])

cluster1 = go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=12,
        line=dict(
            color='rgba(217, 217, 217, 0.14)',
            width=0.5
        ),
        opacity=0.9
    ),
    name="Cluster 0"
)


cluster2 = go.Scatter3d(
    x=x2,
    y=y2,
    z=z2,
    mode='markers',
    marker=dict(
        color='rgb(127, 127, 127)',
        size=12,
        symbol='circle',
        line=dict(
            color='rgb(204, 204, 204)',
            width=1
        ),
        opacity=0.9
    ),
    name="Cluster 1"
)
data2 = [cluster1, cluster2]
layout = go.Layout(
    margin=dict(
        l=0,
        r=0,
        b=0,
        t=30
    )
)

fig = go.Figure(data=data2, layout=layout)

ply.plot(fig, filename='Clusters.html')
ply.plot(fig)
```
