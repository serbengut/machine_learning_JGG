# Linear Support Vector Classifier

```python
# la tecnica mas aplicable de ML para un problema es la maquina de soporte vectorial lineal
# la clasificación o segmentación a traves de mquina de soporte vectorial se puede llevar a cabo con una gran cantidad de opciones:
# el Linear Support Vector Classifier es el más sencillo (SVC)

# veremos un ejemplo muy simple para consolidar qué es trabajar con un SVC.
# el objetivo de una SVC es intentar encajar los datos que le proporcionemos devolviéndonos el mejor hiperplano que sea capaz de dividirlos (de categorizarlos y separarlos)
# a partir de obtener el hiperplano, uno puede nececistar añadir más rasgos o elementos al clasificador para establecer qué tipo de clase predecida nos ha llevado a cabo.
# esto hace que este algoritmo sea muy especifico para nuestro uso cuando tengamos poco datos o se puedan separar de forma lineal, a pesar de que se pueda utilizar en cantidad de situaciones

# vamos con el ejemplo y vamos a establecer las dependencias:
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style    #lo inporto para añadir una visualización de datos
style.use("ggplot")
from sklearn import svm
```

```python
X = [1 , 5 , 1.5 , 8 , 1 , 9]
Y = [2 , 8 , 1.8 , 8 , 0.6 , 11]
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
# a ojo, es más o menos facil de encontrar la linea que separara los tres puntos de la izq de los tres de la derecha
```

```python
# sin embargo, si añadimos más variables (3, 4,...100, 400...) no podremos verlo
# pero la máquina sí puede crear un hiperplano de n-1 dimensiones (n=nº vables.) de manera sencilla
```

```python
# comenzamos a hacer el estudio con estas dos dimensiones:
```

```python
# creamos un dataset:

data = np.array(list(zip(X,Y)))
data
```

```python
# vamos a necesitar cada una de las entradas por razones de entrenamiento
# al ser algoritmo supervisado, necesitamos etiquetar los puntos en 0 y 1, o 1 y 2 según pertenezcan a un grupo o a otro (izquierda o derecha)
```

```python
target = [0, 1, 0, 1, 0, 1]
```

```python
# creamos nuestro clasificador:
    # kernel = tipo de separador: linear: como vamos a buscar un hiperplano, por tanto, un hiperplano en 2D es una recta
    # c = 1.0 (c es una forma de evaluar cómo de mal queremos que sea la clasificación. al final este c es un parametro que tendremos que tunear e ir ajustando)

classifier = svm.SVC(kernel="linear", C = 1.0)
classifier.fit(data, target)
```

```python
# con esto ya hemos podido hacer una separación y entonces la fase de aprendizaje ya está hecha. 
# Para ver la fase de aplicación, a partir de ahora podremos predecir y comprobar si un punto cualquiera entra en la zona de 0 o de 1
```

```python
# probamos con un punto p (con un numero pequeño) y que sabemos que debe entrar en la zona del 0:

p = np.array([0.57, 0.67]).reshape(1,2)
print(p)
classifier.predict(p)
```

```python
# probamos con un punto p (con un numero alto) y que sabemos que debe entrar en la zona del 1:

p = np.array([10.32, 12.67]).reshape(1,2)
print(p)
classifier.predict(p)
```

## Representación gráfica del hiperplano separador en 2D

```python
# vamos a sacar los weights
```

* Modelo: w0 . x + w1 . y + e = 0
* Ecuación del hiperplano en 2D: y = a . x + b 

```python
w = classifier.coef_[0]
w
```

```python
a = -w[0]/w[1]
a
```

```python
b = - classifier.intercept_[0]/w[1]
b
```

```python
xx = np.linspace(0,10)   # (0,10) para englobar todos los puntos en la representación
yy = a * xx + b
```

```python
plt.plot(xx, yy, 'k-', label = "Hiperplano de separación")    
plt.scatter(X, Y, c = target)
plt.legend()
plt.plot()
```
